```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
def custom_resample(wafers,c):
    
    label = (c[:,-1] != 0).astype(int)
    n = len(label)
#     print(Counter(label))
    indices = np.expand_dims(np.arange(n),axis = -1)
    over = RandomOverSampler(sampling_strategy=0.1)
    # fit and apply the transform
    indices_p, label_p = over.fit_resample(indices, label)
    # summarize class distribution
    # define undersampling strategy
    under = RandomUnderSampler(sampling_strategy=0.5)
    # fit and apply the transform
    indices_p, label_p = under.fit_resample(indices_p, label_p)
#     print(Counter(label_p))
    wafers_p = wafers[indices_p[:,0]]
    c_p = c[indices_p[:,0]]
    
    return wafers_p, c_p
```

```python
# inputs = tf.random.normal(shape=(1000, 8, 8))
# labels = tf.random.uniform(shape=(1000, 1), minval=0, maxval=2, dtype=tf.int32)

# other_conditions = tf.random.uniform(shape=(1000, 7), minval=0, maxval=10, dtype=tf.int32)

# cond = tf.concat([labels, other_conditions], axis=1)
train = tf.data.Dataset.load('raw_data_test/all_eLinks_train')
all_dt = list(train)
wafers = []
c = []
for a,b in all_dt:
    wafers.append(a)
    c.append(b)
inputs = np.array(wafers)
cond = np.array(c)
del train
del all_dt

signals_mask = cond[:, -1] > 0
signals = inputs[signals_mask]
signals_cond = cond[signals_mask]
def signal_noise_pairs_generator():
    # Loop through all signal entries
    for signal, signal_cond in zip(signals, signals_cond):
        # Conditions for matching: same values at indices 1 and 2
        matches = tf.logical_and(
            tf.equal(cond[:, 1], signal_cond[1]),
            tf.equal(cond[:, 2], signal_cond[2])
        )
        # Ensure we also select only noise entries
        matches = tf.logical_and(matches, cond[:, -1] < 2)
        
        # Find matching noise inputs
        matching_noise_inputs = inputs[matches]
        
        # Add the first matching noise input to the signal to create an augmented signal
        # Note: If no match is found, return the original signal to handle empty cases
        if tf.shape(matching_noise_inputs)[0] > 0:
            augmented_signal = signal + 0.5*matching_noise_inputs[0]
            # Normalize the augmented signal
            mean, variance = tf.nn.moments(augmented_signal, axes=[0, 1])
            augmented_signal = (augmented_signal - mean) / tf.sqrt(variance)
            
            mean, variance = tf.nn.moments(signal, axes=[0, 1])
            signal = (signal - mean) / tf.sqrt(variance)
        else:
            augmented_signal = signal 
        
        yield signal, augmented_signal

output_types = (tf.float32, tf.float32)  # Assuming the inputs are of type float32
output_shapes = ((8, 8, 1), (8, 8,1))
dataset = tf.data.Dataset.from_generator(
    signal_noise_pairs_generator,
    output_types=output_types,
    output_shapes=output_shapes
)
```

```python
import matplotlib.pyplot as plt

# Function to plot signals with a row-specific color scale
def plot_signals(dataset, num_samples=3):
    fig, axes = plt.subplots(num_samples, 2, figsize=(12, num_samples * 5))

    # Loop through the number of samples required
    for i, (original_signal, augmented_signal) in enumerate(dataset.take(num_samples)):
        if num_samples == 1:
            ax1, ax2 = axes[0], axes[1]
        else:
            ax1, ax2 = axes[i, 0], axes[i, 1]
        
        # Calculate min and max values for this pair of signals
        min_val = min(tf.reduce_min(original_signal).numpy(), tf.reduce_min(augmented_signal).numpy())
        max_val = max(tf.reduce_max(original_signal).numpy(), tf.reduce_max(augmented_signal).numpy())
        
        # Plot original signal
        im1 = ax1.imshow(original_signal.numpy(), cmap='viridis', vmin=min_val, vmax=max_val, interpolation='nearest')
        ax1.set_title("Original Wafer")
        ax1.axis('off')
        
        # Plot augmented signal
        im2 = ax2.imshow(augmented_signal.numpy(), cmap='viridis', vmin=min_val, vmax=max_val, interpolation='nearest')
        ax2.set_title("Augmented Wafer")
        ax2.axis('off')
        
        # Colorbar for this row
        plt.colorbar(im1, ax=[ax1, ax2], orientation='vertical', fraction=0.046, pad=0.04)

#     plt.tight_layout()
    plt.show()

# Now use the function to plot the signals from the dataset
plot_signals(dataset, num_samples=3)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the encoder model
def get_encoder():
    # Create a simple CNN as an encoder that outputs 16D embeddings
    return keras.Sequential([
        keras.Input(shape=(8, 8, 1)),  # Ensure input shape matches the 8x8 data
        layers.Conv2D(16, 3, activation='relu'),
        layers.Flatten(),
        layers.Dense(16, activation='relu'),
        layers.Dense(4, activation='relu')  # Output layer directly encoding to 16D space
    ])

# Define the contrastive model using model-subclassing
class ContrastiveModel(keras.Model):
    def __init__(self, temperature=0.1):
        super().__init__()
        self.temperature = temperature
        self.encoder = get_encoder()

        # Print model summary
        self.encoder.summary()

    def compile(self, optimizer, **kwargs):
        super().compile(**kwargs)
        self.optimizer = optimizer
        self.contrastive_loss_tracker = keras.metrics.Mean(name="c_loss")

    @property
    def metrics(self):
        return [self.contrastive_loss_tracker]

    def contrastive_loss(self, features_1, features_2):
        # Normalize features
        features_1 = tf.math.l2_normalize(features_1, axis=1)
        features_2 = tf.math.l2_normalize(features_2, axis=1)
        # Compute dot product, apply temperature scaling
        similarities = tf.matmul(features_1, features_2, transpose_b=True) / self.temperature
        # Create labels for the positive pairs
        batch_size = tf.shape(features_1)[0]
        labels = tf.range(batch_size)
        labels = tf.reshape(labels, [batch_size, 1])
        # Create mask for positive and negative pairs
        mask = tf.math.equal(labels, tf.transpose(labels))
        mask = tf.cast(mask, tf.float32)

        # Compute contrastive loss using the masked softmax cross-entropy
        logits = tf.math.exp(similarities)
        positive_samples = tf.reduce_sum(mask * logits, axis=1)
        all_samples = tf.reduce_sum(logits, axis=1)
        loss = -tf.math.log(positive_samples / all_samples)
        return tf.reduce_mean(loss)

    def train_step(self, data):
        # Unpack data. Each batch will contain pairs of augmented data
        x1, x2 = data
        with tf.GradientTape() as tape:
            features_1 = self.encoder(x1, training=True)
            features_2 = self.encoder(x2, training=True)
            loss = self.contrastive_loss(features_1, features_2)

        gradients = tape.gradient(loss, self.encoder.trainable_weights)
        self.optimizer.apply_gradients(zip(gradients, self.encoder.trainable_weights))
        self.contrastive_loss_tracker.update_state(loss)

        return {m.name: m.result() for m in self.metrics}

    def test_step(self, data):
        x1, x2 = data
        features_1 = self.encoder(x1, training=False)
        features_2 = self.encoder(x2, training=False)
        loss = self.contrastive_loss(features_1, features_2)
        self.contrastive_loss_tracker.update_state(loss)

        return {m.name: m.result() for m in self.metrics}
```

```python
train_dataset = dataset.take(2500).batch(100)

# Create the test dataset
test_dataset = dataset.skip(2500).take(500).batch(100)
# Instantiate the simplified contrastive model
pretraining_model = ContrastiveModel()

# Compile the model with just the contrastive optimizer
pretraining_model.compile(
    optimizer=keras.optimizers.Adam()  # Use a single optimizer for the model
)

# Fit the model to the training data and validate using the test dataset
pretraining_history = pretraining_model.fit(
    train_dataset,  # Assume train_dataset is properly formatted with pairs
    epochs=50,
    validation_data=test_dataset  # Assume test_dataset is also formatted with pairs
)

# Since we removed the probe and its accuracy, let's adjust the print statement
# to reflect only the loss or whatever metrics we defined
print(
    "Final validation loss: {:.2f}".format(
        pretraining_history.history["c_loss"][-1]
    )
)
```

```python
import tensorflow as tf
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming test_dataset is already prepared as described
# Assuming the model is named 'pretraining_model' and the encoder is accessible

# Extract latent features and corresponding 'cond' labels from the signal samples
train = tf.data.Dataset.load('raw_data_test/all_eLinks_test').batch(100)
all_dt = list(train)[0:2000]
conds = []
latent_features = []
for inputs, cond in all_dt:
    
    # Get latent features from the encoder
    latent = pretraining_model.encoder(inputs, training=False)
    latent_features.append(latent.numpy())
    conds.append(cond.numpy())

latent_features = np.concatenate(latent_features, axis=0)
conds = np.concatenate(conds, axis=0)

# Normalize features for PCA
scaler = StandardScaler()
latent_features_normalized = scaler.fit_transform(latent_features)

# Perform PCA to reduce to two dimensions
tsne = TSNE(n_components=2, perplexity=30, learning_rate='auto', init='random')
latent_features_tsne = tsne.fit_transform(latent_features_normalized)

# pca = PCA(n_components=2)
# latent_features_pca = pca.fit_transform(latent_features_normalized)
```
